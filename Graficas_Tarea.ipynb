```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# @title <h2><b> Pregunta 3 </b><h2>
y0 = 1
v0 = 0
dseta = 0
wn = 1
wd = wn*np.sqrt(1-dseta**2)
A1 =  y0
A2 = (v0 + dseta*wn*y0)/wd

t = np.linspace(0,100,10000)
y = np.exp(-dseta*wn*t)*(A1*np.cos(wd*t)+A2*np.sin(wd*t))

# Grfica de Posicin
plt.plot(t, y, '-k', label = "$\zeta$ = 0")
plt.xlabel("tiempo $[s]$")
plt.ylabel("$y$ [m]")
plt.xlim([0,50])
plt.legend()
plt.show()

y0 = 1
v0 = 0
dseta = 0.02
wn = 1
wd = wn*np.sqrt(1-dseta**2)
A1 =  y0
A2 = (v0 + dseta*wn*y0)/wd

t = np.linspace(0,100,10000)
y = np.exp(-dseta*wn*t)*(A1*np.cos(wd*t)+A2*np.sin(wd*t))

# Grfica de Posicin
plt.plot(t, y, '-k', label = "$\zeta$ = 0.02")
plt.xlabel("tiempo $[s]$")
plt.ylabel("$y$ [m]")
plt.xlim([0,50])
plt.legend()
plt.show()
```

```python
# @title <h2><b> Pregunta 4 </b><h2>
X = 1
Omega = 5
y0 = 1
v0 = 0
dseta = 0
wn = 1
wd = wn*np.sqrt(1-dseta**2)
A1 =  y0
A2 = (v0 + dseta*wn*y0-X*Omega)/wd

t = np.linspace(0,100,10000)
y = np.exp(-dseta*wn*t)*(A1*np.cos(wd*t) + A2*np.sin(wd*t) + X*np.sin(Omega*t))

# Grfica de Posicin
plt.plot(t, y, '-k', label = "$\zeta$ = 0")
plt.xlabel("tiempo $[s]$")
plt.ylabel("$y$ [m]")
plt.xlim([0,50])
plt.legend()
plt.show()

y0 = 1
v0 = 0
dseta = 0.02
wn = 1
wd = wn*np.sqrt(1-dseta**2)
A1 =  y0
A2 = (v0 + dseta*wn*y0-X*Omega)/wd

t = np.linspace(0,100,10000)
y = np.exp(-dseta*wn*t)*(A1*np.cos(wd*t)+A2*np.sin(wd*t)+X*np.sin(Omega*t))

# Grfica de Posicin
plt.plot(t, y, '-k', label = "$\zeta$ = 0.02")
plt.xlabel("tiempo $[s]$")
plt.ylabel("$y$ [m]")
plt.xlim([0,50])
plt.legend()
plt.show()
```

```python
# @title <h2><b> Pregunta 5 </b><h2>

r = np.linspace(0,4,10000)

def factor_amplificacion(dseta):
  return 1/np.sqrt( (1-r**2)**2 + (2*dseta*r)**2 )

G1  = factor_amplificacion(0)
G2  = factor_amplificacion(0.1)
G3  = factor_amplificacion(0.2)
G4  = factor_amplificacion(0.3)
G5  = factor_amplificacion(0.4)
G6  = factor_amplificacion(0.5)
G7  = factor_amplificacion(0.6)
G8  = factor_amplificacion(0.7)
G9  = factor_amplificacion(0.8)
G10 = factor_amplificacion(0.9)
G11 = factor_amplificacion(1.0)
G12 = factor_amplificacion(1.1)
G13 = factor_amplificacion(1.2)
G14 = factor_amplificacion(1.3)
G15 = factor_amplificacion(1.4)
G16 = factor_amplificacion(1.5)
G17 = factor_amplificacion(1.6)
G18 = factor_amplificacion(1.7)
G19 = factor_amplificacion(1.8)
G20 = factor_amplificacion(1.9)
G21 = factor_amplificacion(2.0)


# Grfica de Posicin
plt.plot(r, G1,  '-k')
plt.plot(r, G2,  '-k')
plt.plot(r, G3,  '-k')
plt.plot(r, G4,  '-k')
plt.plot(r, G5,  '-k')
plt.plot(r, G6,  '-k')
plt.plot(r, G7,  '-k')
plt.plot(r, G8,  '-k')
plt.plot(r, G9,  '-k')
plt.plot(r, G10, '-k')
plt.plot(r, G11, '-k')
plt.plot(r, G12, '-k')
plt.plot(r, G13, '-k')
plt.plot(r, G14, '-k')
plt.plot(r, G15, '-k')
plt.plot(r, G16, '-k')
plt.plot(r, G17, '-k')
plt.plot(r, G18, '-k')
plt.plot(r, G19, '-k')
plt.plot(r, G20, '-k')
plt.plot(r, G21, '-k')
plt.xlabel("r")
plt.ylabel("G")
plt.xlim([0,4])
plt.ylim([0,6])
plt.show()

# Grfica de Posicin
plt.plot(r, G1,  '-k', label = "$\zeta$ = 0")
plt.plot(r, G2,  '-r', label = "$\zeta$ = 0.1")
plt.plot(r, G3,  '-b', label = "$\zeta$ = 0.2")
plt.plot(r, G6,  '-g', label = "$\zeta$ = 0.5")
plt.plot(r, G8,  '-y', label = "$\zeta$ = 0.7")
plt.plot(r, G11, '-', label = "$\zeta$ = 1.0")
plt.plot(r, G16, '-' , label = "$\zeta$ = 1.5")
plt.plot(r, G19, '-', label = "$\zeta$ = 1.8")
plt.plot(r, G21, '-', label = "$\zeta$ = 2.0")
plt.xlabel("r")
plt.ylabel("G")
plt.xlim([0,4])
plt.ylim([0,6])
plt.legend()
plt.show()
```

```python
# @title <h2><b> Pregunta 6 </b><h2>

r_data = np.linspace(0,4,1000)

def angulo_desfase(dseta):
  gamma = []
  for r  in np.linspace(0,4,1000):
    if r<= 1:
      gamma.append( np.arctan( (2*dseta*r) / (1-r**2) ) )
    else: 
      gamma.append( np.arctan( (2*dseta*r) / (1-r**2) ) + np.pi )

  return np.array(gamma)

gamma1  = angulo_desfase(0.0)
gamma2  = angulo_desfase(0.1)
gamma3  = angulo_desfase(0.2)
gamma4  = angulo_desfase(0.3)
gamma5  = angulo_desfase(0.4)
gamma6  = angulo_desfase(0.5)
gamma7  = angulo_desfase(0.6)
gamma8  = angulo_desfase(0.7)
gamma9  = angulo_desfase(0.8)
gamma10 = angulo_desfase(0.9)
gamma11 = angulo_desfase(1.0)
gamma12 = angulo_desfase(1.1)
gamma13 = angulo_desfase(1.2)
gamma14 = angulo_desfase(1.3)
gamma15 = angulo_desfase(1.4)
gamma16 = angulo_desfase(1.5)
gamma17 = angulo_desfase(1.6)
gamma18 = angulo_desfase(1.7)
gamma19 = angulo_desfase(1.8)
gamma20 = angulo_desfase(1.9)
gamma21 = angulo_desfase(2.0)

#label = "$\zeta$ = 0"
plt.plot(r_data, gamma1,  '--', label = "$\zeta$ = 0" )
plt.plot(r_data, gamma2,  '--', label = "$\zeta$ = 0.1" )
plt.plot(r_data, gamma3,  '--', label = "$\zeta$ = 0.2" )
plt.plot(r_data, gamma4,  '--' )
plt.plot(r_data, gamma5,  '--' )
plt.plot(r_data, gamma6,  '--', label = "$\zeta$ = 0.5" )
plt.plot(r_data, gamma7,  '--' )
plt.plot(r_data, gamma8,  '--' )
plt.plot(r_data, gamma9,  '--' )
plt.plot(r_data, gamma10,  '--' )
plt.plot(r_data, gamma11,  '--', label = "$\zeta$ = 1.0" )
plt.plot(r_data, gamma12,  '--' )
plt.plot(r_data, gamma13,  '--' )
plt.plot(r_data, gamma14,  '--', label = "$\zeta$ = 1.3" )
plt.plot(r_data, gamma15,  '--' )
plt.plot(r_data, gamma16,  '--', label = "$\zeta$ = 1.5")
plt.plot(r_data, gamma17,  '--' )
plt.plot(r_data, gamma18,  '--' )
plt.plot(r_data, gamma19,  '--', label = "$\zeta$ = 1.8" )
plt.plot(r_data, gamma20,  '--' )
plt.plot(r_data, gamma21,  '--k', label = "$\zeta$ = 2.0" )

plt.xlabel("r")
plt.ylabel("$\gamma$ [rad]")
plt.xlim([0,4])
plt.ylim([0,np.pi])
plt.legend()
plt.show()


```
